```python
import datetime
import sys
from deep_translator import GoogleTranslator
import pandas
```

```python


if __name__ == '__main__':
    df = pandas.read_csv("parkman.csv", sep=";",engine='python')
    df['content'].replace([';'],'',inplace=True)

    for i in df.index:
        try:
            translated = GoogleTranslator(source='auto', target='en').translate(df.iloc[i]["content"])
            df.iloc[i, 2] = translated
            print(df.iloc[i]["ind"], translated)
            print("---------------------------------------------------------------------------------")
            with open('/Users/WorkSpace/Desktop/easyparkhalf1_trans.csv', 'a',encoding='utf-8') as fd:
                user = df.iloc[i]["userName"]
                rating = df.iloc[i]["rating"]
                date = df.iloc[i]["date"]
                version = df.iloc[i]["reviewVersion"]
                fd.write(str(i)+";"+str(user)+";"+str(translated)+";"+str(rating)+";"+str(date)+";"+str(version)+"\n")

        except:
            print("*********************************WARNING*************************************************")
            print("Oops!", sys.exc_info()[0], "occurred on.", df.iloc[i]["content"], df.iloc[i]["ind"])
            print("*********************************WARNING*************************************************")
            df.iloc[i, 2] = "Could not translate"
            pass

           
```
